# Feature Engineering

Applying Feature engineering to extract the best features.
 First, we use univariate feature selection.
Second, we use the recursive feature elimination.

```python
# Feature Extraction with Univariate Statistical Tests (Chi-squared for classification)
from pandas import read_csv
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

array = data.values
X = array[:,0:-1]
Y = array[:,-1]
# feature extraction
test = SelectKBest(score_func=chi2, k=4)
fit = test.fit(X, Y)
# summarize scores
set_printoptions(precision=3)
print(fit.scores_)
features = fit.transform(X)
```

```python
# Feature Extraction with RFE
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
# feature extraction
model = LogisticRegression(max_iter=400)
rfe = RFE(model)
fit = rfe.fit(X, Y)
```

```python
fit.n_features_
```

```python
fit.support_
```

```python
fit.ranking_
```

**After feature extraction using univariate feature selection and recursive feature elimination models. We arrive at the conclusion that 3 features- competetiveness, financial_flexibility and credibility are the most important**

# Model Building Logistic Regression

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
model_logistic=LogisticRegression()
```

```python
model_logistic.fit(x_train,y_train)
```

```python
#predicting the class of the test data points
pred_logistic=model_logistic.predict(x_test)
```

```python
pred_logistic
```

Importing confusion matrix to check the true postive and true negative values of the predicted and testing values

```python
from sklearn.metrics import confusion_matrix
confusion_logistic=confusion_matrix(y_test,pred_logistic)
confusion_logistic
```

```python
from sklearn.metrics import accuracy_score
acc_logistic=accuracy_score(y_test,pred_logistic)
print("The accuracy of the model on the testing dataset is:",acc_logistic*100,"%")
```

```python
pred_logistic
```

Running the model on the entire dataset to check the accuracy score.

```python
pred_logistic_entire=model_logistic.predict(x)
confusion_logistic_entire=confusion_matrix(y,pred_logistic_entire)
confusion_logistic_entire
```

```python
print("The accuracy obtained by the model after predicting values of the entire dataset:",accuracy_score(y,pred_logistic_entire)*100,"%")
```

```python
print("the accuracy obtained by the model on the training data set:",model_logistic.score(x_train,y_train)*100,"%")
```

```python
print("the accuracy obtained by the model on the testing data set:",model_logistic.score(x_test,y_test)*100,"%")
```

```python
print(classification_report(y_test,pred_logistic))
```

**From the above accuracy scores obtained, we notice that the testing accuracy is higher than or equal to the training accuray, which is not consistent with classification models in case of a proper learning. We can conclude that the model is overfitted.**

**I will be applying the ridge, lasso and elastic net regression to see which is the best regularization model**

Ridge Regularization

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=10, max_iter=100, tol=0.1)#after multiple trial and run methods alpha value is chosen as 10
ridge_reg.fit(x_train, y_train)
pred_ridge=ridge_reg.predict(x_test)
```

```python
print("The accuracy obtained by Ridge regularization on the testing dataset is :",ridge_reg.score(x_test,y_test)*100,"%")
```

```python
print("The accuracy obtained by Ridge regularization on the training dataset is :",ridge_reg.score(x_train, y_train)*100,"%")
```

As the Ridge model provides a higher accuracy for the training dataset than testing dataset, we can use this model for the final deployment
